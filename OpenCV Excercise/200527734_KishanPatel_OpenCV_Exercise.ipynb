```python
#read in the detect_blob image
#then draw a box shape around the blue solid box in the image that is your region of interest
#crop your region of interest
#and then rotate it 45 degrees so that it is not clipped

#Answer
# Open the image file
blobImage = cv2.imread("detect_blob.png")

# Display the image
cv2.imshow('Blob Image', blobImage)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Define the region of interest (ROI) as a rectangle

roi = cv2.rectangle(blobImage, (365, 460), (470, 570), (0, 0, 255), 2)

# Display the image with the ROI drawn on it
cv2.imshow('Image with ROI', roi)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Crop the region of interest from the image
cropped_image = blobImage[460:570, 365:470]

# Display the cropped image
cv2.imshow('Cropped Image', cropped_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Define the rotation angle in degrees
angle = 45

# Get the dimensions of the cropped image
h, w = cropped_image.shape[:2]

# Define the rotation matrix
M = cv2.getRotationMatrix2D((w/2, h/2), angle, 1)

# Apply the rotation to the cropped image
rotated_image = cv2.warpAffine(cropped_image, M, (w, h))

# Display the rotated image
cv2.imshow('Rotated Image', rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#read in the thresh image, perform thresholding by using gaussian blur with thresholds 5,55
#then perform dilation, erosion, opening & closing using a 5x5 kernel with 1 iteration on the original thresh image (Hint: look into Opencv for operations)
#Finally apply canny edge detection - experiment using wide and narrow thresholds 
#(so just 2 images in total one wide and one narrow)

#Answer
# read in the image
threshold_ = cv2.imread("thresh.jpg")

# apply Gaussian blur with a kernel size of 5
blur5 = cv2.GaussianBlur(threshold_, (5,5), 0)

# apply thresholding with a threshold value of 55
ret, thresh = cv2.threshold(blur5, 55, 255, cv2.THRESH_BINARY)

# display the thresholded image
cv2.imshow("Original Image", threshold_)
cv2.imshow("Thresholded Image", thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# create a 5x5 kernel
kernel = np.ones((5, 5), np.uint8)

# perform dilation with 1 iteration
dilation = cv2.dilate(threshold_, kernel, iterations=1)

# perform erosion with 1 iteration
erosion = cv2.erode(threshold_, kernel, iterations=1)

# perform opening with 1 iteration
opening = cv2.morphologyEx(threshold_, cv2.MORPH_OPEN, kernel, iterations=1)

# perform closing with 1 iteration
closing = cv2.morphologyEx(threshold_, cv2.MORPH_CLOSE, kernel, iterations=1)

# display the results
cv2.imshow("Original", threshold_)
cv2.imshow("Dilation", dilation)
cv2.imshow("Erosion", erosion)
cv2.imshow("Opening", opening)
cv2.imshow("Closing", closing)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# apply Canny edge detection with a wide threshold
canny_wide = cv2.Canny(threshold_, 50, 150)

# apply Canny edge detection with a narrow threshold
canny_narrow = cv2.Canny(threshold_, 100, 120)

# display the results
cv2.imshow("Wide Threshold", canny_wide)
cv2.imshow("Narrow Threshold", canny_narrow)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#create a script that opens the picture and allows you to draw empty red circles where ever you click the RIGHT MOUSE BUTTON DOWN

#Answer
# Define the callback function for mouse events
def draw_circle(event, x, y, flags, param):
    global img

    if event == cv2.EVENT_RBUTTONDOWN:
        # Draw a circle at the mouse position
        cv2.circle(img, (x, y), 20, (0, 0, 255), 2)

        # Display the updated image
        cv2.imshow('image', img)

# Load the image
img = cv2.imread('players.jpg')

# Display the image and set the mouse callback function
cv2.imshow('image', img)
cv2.setMouseCallback('image', draw_circle)

# Wait for a key event
while True:
    key = cv2.waitKey(1) & 0xFF

    # If the 'q' key is pressed, exit the loop
    if key == ord('q'):
        break

# Close the window and release resources
cv2.destroyAllWindows()
```

```python
#open and display the any image of your choice
#apply a binary threshold onto the image
#convert image colorspace to HSV and display the image
#create a low pass filter with a 4 by 4 Kernel filled with values of 1/10 (0.01) and then use 2-D Convolution to blur the image (displayed in normal RGB)
#create a Horizontal Sobel Filter with a kernel size of 5 to the grayscale version of the image and then display the resulting gradient filtered version of the image.
#plot the color histograms for the RED, BLUE, and GREEN channel of the image. Pay careful attention to the ordering of the channels

#Answer
# Load the image
img = cv2.imread('players.jpg')

# Display the original image
cv2.imshow('Original', img)
cv2.waitKey(0)

# Apply binary thresholding
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

# Display the binary image
cv2.imshow('Thresholded', thresh)
cv2.waitKey(0)

# Convert to HSV colorspace
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Display the HSV image
cv2.imshow('HSV', hsv)
cv2.waitKey(0)

# Create low pass filter
kernel = np.ones((4,4),np.float32)/16

# Apply 2D convolution with low pass filter
blur = cv2.filter2D(img,-1,kernel)

# Display the blurred image
cv2.imshow('Blurred', blur)
cv2.waitKey(0)

# Apply Sobel filter to grayscale image
sobelx = cv2.Sobel(gray,cv2.CV_64F,1,0,ksize=5)

# Display the gradient filtered image
cv2.imshow('Gradient', sobelx)
cv2.waitKey(0)

# Calculate and plot color histograms
colors = ('b', 'g', 'r')
for i, col in enumerate(colors):
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.fill_between(np.arange(256), hist.ravel(), 0, color=col)
    plt.xlim([0, 256])

plt.show()

# Close all windows
cv2.destroyAllWindows()
```
